```python

import pandas as pd 
import numpy as np
```

```python
def logCPM(a):
    id1=a.iloc[:,0]
    b=a.select_dtypes(include=['int64'])
    a=a._get_numeric_data()
    a=a.values
    try:
        cpm=[]
        for i in range(0,len(a)):
                colsum=a[:,i].sum()

                c1=(a[:,i]/colsum)*10**6
                cpm.append(c1)
    
    except:
        print("ignore error")
    cpm=np.array(cpm)

    lgv=np.log2(cpm+1) 
    logcpm=pd.DataFrame(lgv).T
    logcpm.columns=b.columns
    logcpm['Geneid']=id1
    lk=logcpm.set_index('Geneid')
    return lk
    
```

```python
data=pd.read_csv('GSE184807_Raw_count_data.txt.gz',sep="\t")
data
```

```python
#calling function logCPM
logCPM(data)
```
